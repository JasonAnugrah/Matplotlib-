```python
Import Modules
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
Sample dataset
```

```python
data = {'Item A' : 109438,
        'Item B' : 103569,
        'Item C' : 112214,
        'Item D' : 112591,
        'Item E' : 100934,
        'Item F' : 103660,
        'Item G' : 137351,
        'Item H' : 123381,
        'Item I' : 135841,
        'Item J' : 104437}
data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

```python
Simple plot
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

```python
Pengaturan style
```

```python
plt.style.available
```

```python
plt.style.use('bmh')

fig, ax = plt.subplots()
ax.barh(items,count)
plt.show()
```

```python
Pengaturan tick label
```

```python
fig, ax = plt.subplots()
ax.barh(items,count)

labels = ax.get_xticklabels()

#set properties
plt.setp(labels,
         rotation=45,
         horizontalalignment='right')

plt.show()
```

```python
Pengaturan format pada ticker
```

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

```python
Pengaturan label pada sumbu (axis) dan judul (title)
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

```python
Penambahan garis (vertical/horizontal line) pada plot
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
           ls='--',
           color='r')

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

```python
Menyimpan hasil plot ke dalam suatu file
```

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
           ls='--',
           color='r')

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

fig.savefig('sales.png',
            transparent=False,
            dpi=80)
```
